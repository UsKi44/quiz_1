```python
first_col_name = "YearsCode"

df[first_col_name].replace({"More than 50 years": 51, "Less than 1 year":0}, inplace=True)

df[first_col_name].replace({"NA":99}, inplace=True)

df[first_col_name] = pd.to_numeric(df[first_col_name])

first_plt = plt.figure(figsize=(20,20))
second_plt = plt.figure(figsize=(20,20))

sns.barplot(x=df[first_col_name].value_counts().index, y= df[first_col_name].value_counts().values)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Frequency of Values in the column")

first_plt.show()
```

```python
second_col_name = "YearsCodePro"

df[second_col_name].replace({"More than 50 years": 51,"Less than 1 year": 0}, inplace= True)


df[second_col_name].replace({"NA": 99}, inplace=True)

df[second_col_name] = pd.to_numeric(df[second_col_name])

sns.barplot(x=df[second_col_name].value_counts().index, y= df[second_col_name].value_counts().values)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Frequency of Values in the column")
```
